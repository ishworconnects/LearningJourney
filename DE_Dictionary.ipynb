A dictionary is a collection used to store data as key-value pairs, where each key is used to get its value.

```python
this_dict = {"name": "John", "age": 30, "city": "New York"}
print(this_dict)
```

```python
print(this_dict["name"])
```

```python
this_data = {
    "name": "John",
    "class": "BDE",
    "year": 2024
}
print(this_data)
```

```python
print(this_data["name"])
print(this_data["class"])
```

```python
student = {
    "name": "Matrika",
    "year": 2024,
    "course": "DE"
}
print(student)
```

Key: a key is like a label or name used to identify the data.

```python
student['name']
```

We use the key inside `[]` to get the value associated with that key.

```python
student = {
    "name": "Iswor",
    "year": 2025,
    "course": "DE"
}
print(student)
```

```python
student['course']
```

```python
print(student.get("city", "City not found"))
```

```python
student["city"] = "Helsinki"
print(student.get("city"))
```

`get()` is a safe way to check for a key because it does not raise an error when the key is missing.

```python
student.pop('year')
```
