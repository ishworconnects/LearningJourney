map()
  |
  |----> Built-in Python function
        |
        |----> Applies a function
              to every element
              of a sequence, like a list

map(function,iterable)
#function -logic to apply
#iterable -> list/data

#widely used in data pipelines

```python
# Add 10 to each number using for loop

numbers = [1, 2, 3, 4, 5]
result = []

for num in numbers:
    result.append(num + 10)

print(result)
```

```python
# Map functions
 
number = [1,2,3,4,5]
 
def add_ten(x):  # function defined
    return x + 10   # Logic applied
 
result = map(add_ten, number)  # applies function to each element
 
print(list(result)) # convert result into list
```

```python
# Add 10% bonus using map function

sal = [10000, 20000, 30000]

def add_bonus(salary):
    return salary + (salary * 0.10)

result = map(add_bonus, sal)

print(list(result))
```

```python
sal = [10000, 20000, 30000]

def add_bonus(salary):
    return salary + (salary * 0.10)

result = map(add_bonus, sal)

for salary in result:
    print("Salary after 10% bonus =", salary)
```

```python
# lambda with map
numbers = [1, 2, 3]
result = map(lambda x: x + 10, numbers)
print(list(result))
```

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

res = map(lambda x, y: x + y, list1, list2)
print(list(res))
```

Filter -built in python function used to select elements from a dataset based on a condition
it keeps only those values where the condition is TRUE
apply condition -> keep valid data -> remove unwanted data

#filter(function(condition- returnTrue /False)iterable(list/dataset))

#map- transform data
#filter()- select data
#lambda()->define logic

```python
#get only even number

num = [1, 2, 3, 4, 5, 6]
result = []

for n in num:
    if n % 2 == 0:
        result.append(n)

print(result)
```

```python
#use filter
num = [1,2,3,4,5,6]

result1 = filter(lambda x: x%2 ==0,num)
#filter returns an iterator
print(list(result1))
```

```python
# get employees with salary > 10000

s = [5000, 12000, 8000, 20000, 50000]

high_sal = filter(lambda x: x > 10000, s)

print(list(high_sal))
```

fibonacci series 

Sequence where  0,1,1,2,3,5,8,13

f(n)=f(n-1)+f(n-2)

```python
def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1)+fib(n-2)

print(fib(5))
```

```python
#f(1)=1 , f(0)=0
```

```python
#fibonacci Series

def fib(n):
    if n<= 1:
        return n
    return fib(n-1) + fib(n-2)

#print series
n = 7
for i in range(n):
    print(fib(i))
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(8))
```
