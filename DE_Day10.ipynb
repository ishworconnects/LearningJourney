```python
Introduction to Hadoop and Its Architecture

Data warehousing is the process of collecting, storing, and managing data from different sources in a centralized repository.

Key point:
It consolidates data from multiple sources and helps businesses track historical data and trends.
```

```python
Hadoop – Open-Source Framework

Stores massive volumes of data
Processes data efficiently
Works across multiple computers instead of a single machine
Helps organizations manage big data by distributing work across many systems
```

```python
Why Hadoop Is Needed

Data comes from social media, e-commerce, banking, IoT devices, and mobile apps
The volume of data is very large
Data must be processed quickly
Data exists in many different formats
```

```python
Hadoop in a Distributed Environment

Splits large data into smaller blocks
Stores data across many computers
Processes data in parallel
Improves speed and efficiency in handling big data
```

```python
Cluster in Hadoop

A cluster is a group of connected computers
These computers work together as one system
Hadoop runs on clusters to store and process large amounts of data
```

```python
Hadoop Scalability-

it can scale from 1 machine to 1000+ machines

if data increases:

just add more computers to the cluster
no need to replace the entire System
```

```python
3 V’s of Big Data

Volume – huge amounts of data, such as TB, PB, and EB
Velocity – the speed at which data is generated, such as real-time stock trades and live streams
Variety – different types of data formats
```
