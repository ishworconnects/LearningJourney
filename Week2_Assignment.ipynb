Q6. Low Sales Alert
Display products where Sales < 10

```python
#Q1. Create the DataFrame
import pandas as pd

data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones'],
    'Price':   [75000, 800, 1500, 22000, 3500],
    'Sales':   [12, 45, 30, 8, 25]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Q2. Add Revenue column
df['Revenue'] = df['Price'] * df['Sales']
print(df)
```

```python
#Q3. Basic Operations

print("\nFirst 3 rows:")
print(df.head(3))
```

```python
print("\nColumn names:")
print(df.columns)
```

```python
print("\nData types:")
print(df.dtypes)
```

```python
#Q4. Add a new record using loc
df.loc[len(df)] = ['Webcam', 2500, 15, 2500 * 15]
print(df)
```

```python
#Q5. Missing Data Handling
import numpy as np

# Add missing values manually
df.loc[1, 'Sales']   = np.nan
df.loc[3, 'Price']   = np.nan
df.loc[5, 'Revenue'] = np.nan

# Detect missing values /Data Frame with missing values
print(df.isnull())

# Count missing values per column
print(df.isnull().sum())

# Fill missing values with 0
df = df.fillna(0)
print(df)
```

```python
#Q6. Low Sales Alert
low_sales = df[df['Sales'] < 10]
print(low_sales)
```
