1. Can keys be numbers?
Step 1: In Python, dictionary keys must be hashable.
Step 2: Numbers are hashable.
Step 3: So, yes, numbers can be used as dictionary keys.

```python
my_data1 = {
    1:"One"
    2:"Two"
}
print(my_data1[1])
```

```python
d = {1: "apple", 2: "banana"}
print(d)
```

2. Can keys be a list [1, 2]?
Step 1: Dictionary keys must be hashable.
Step 2: Lists are not hashable because they can be changed.
Step 3: So, no, a list cannot be used as a dictionary key.

```python
my_data = {
    [1, 2, 3]: "Value"
}
print(my_data)# Keys should be immutable, so we can not assign any mutable collection or data structure as a key in a dict
```

```python
d = {[1, 2]: "fruit"}   # Error
```

```python
d = {(1, 2): "fruit"}
print(d)
```

Number as key → Yes
List as key → No
