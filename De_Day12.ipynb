INNER JOIN

```python
%sql
SELECT e.emp_id, e.emp_name, d.department_name
from employee e
INNER JOIN departments d
ON e.department_id = d.department_id;
```

get all employees even if they don't  have department

```python
%sql
--all records from left table and matching records from right table
SELECT e.emp_id, e.emp_name, d.department_name
from employee e
INNER JOIN departments d
ON e.department_id = d.department_id;
```

get all departments even if no employees exists

```python
%sql
--get all departments even if no employees exists
SELECT e.emp_id, e.emp_name, d.department_name
from employee e
RIGHT JOIN departments d
ON e.department_id = d.department_id;
```

RIGHT -JOIN - get all employees and all departments- (LEFT + RIGHT)

```python
%sql
SELECT e.emp_name, d.department_name
from employee e
CROSS JOIN departments d;
```

which function should be used for top N per group
-- Row_number() - unique, consective, numbering (pagination)

-- rank() - to rank items where ties skip intermediate rank (1,2,2,4)

-- dense_rank() - for ranking without gaps (1,2,2,3) 
-- dense_rank() - for ranking without gaps (1,2,2,3) -
-- assign the same rank to ties without gaps (1,2,2,3)
-- Leaderboards -> where ranking must be contiguous, identifying top N products
 
 
e.g.,
     rn
100  1  1  1
100  2  1  1
80   3  3  2
70   4  4  3

```python
%sql
-- Row_number - Unique rank(no duplicates)
-- Rank() - skips rank() - 1,2,2,4
-- DENSE_RANK() - No skipping - no skipping (1,2,2,3)
select * ,
row_number() over(partition by department order by salary desc) as rank,
rank() over(partition by department order by salary desc) as rank_val,
dense_rank() over(partition by department order by salary desc) as dense_rank_val
from emp1;
```

write a query to calculate the running total of salries ordered by employee ID

```python
%sql
SELECT
    emp_id,
    salary,
    SUM(salary) OVER (ORDER BY emp_id) AS running_total_salary
FROM emp1;
```
