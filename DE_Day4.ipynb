Is alpha() function - checks wheter all character in a string are alphabets(letters only)
string.isalpha()it returns a boolean value - True- all are chars(A-Z and a-z), False- if any character is not a letter

```python
text = "Bipina"
my_text = "Nick123"
my_text1 = ""
my_text2= "hii Coders"
my_text3= "Hello!"
print(text.isalpha())
print(my_text1.isalpha())
print(my_text2.isalpha())
print(my_text3.isalpha())
```
