SET - collection that stores unique values only, 
doesnot allow duplicates and does not maintain order

creates using {}

```python
my_set = {1, 2,2,3,3, 5,4,6,4,8}
my_set

#we addded duplicates(2,3,4) but they are not stored in the set because sets only store unique elements

#python stores value in a way that - checks if a value is already exists in the set, if it does not exist, it adds the value to the set, if it already exists, it ignores the value and does not add it to the set. This is why we only have unique values in a set.
#unordered 
```

```python
my_set={1, 2, 3, 4, 5, 6, 8}
my_set.add(7) #add method to add an element to the set  
print(my_set)

#If you try to remove the any number not present on the set it will give key error

my_set.remove(10) #remove method to remove an element from the set, if the element is not present it will raise a KeyError
print(my_set)

```

```python
my_set.discard(10) #discard method to remove an element from the set, if the element is not present it will not raise a KeyError
print(my_set)
```

```python
len(my_set) #length of the set
```

```python
max(my_set) #maximum value in the set
```

```python
min(my_set) #minimum value in the set
```

```python
sum(my_set) #sum of all the elements in the set
```

```python
#2-Add Elements to a Set
#add value 10 to the set
my_set = {1, 2, 3}
my_set.add(10)
print(my_set)   
```

```python
#3-Remove Elements safely-remove element from which is not present in the set without raising an error
my_set = {1, 2, 3, 10}
my_set.add(4)   # add an element to the set
print(my_set)
my_set.discard(15)   # does nothing if 15 is not present
print(my_set)#discard method to remove an element from the set, if the element is not present it will not raise a KeyError
```

```python
#Find common elements between two sets
A = {1, 2, 3, 4, 5}
B= {4, 5, 6, 7, 8}
common_elements = A & B #intersection of sets A and B
print(common_elements)
```

```python
A= {1, 2, 3, 4}
B= {3, 4, 7, 8}

diff = A^B #symmetric difference of sets A and B (elements in A or B but not in both)
print(diff)

#combine two sets to find unique elements
union = A | B #union of sets A and B
print(union)
```
